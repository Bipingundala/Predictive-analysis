```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset directly from URL (public dataset)
url = "/content/loan_approval_dataset.csv"
df = pd.read_csv(url)

# Show first few rows
print(df.head())
print(df.columns)
# Data cleaning & preprocessing

# Strip leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# Convert target variable to numeric (if necessary)
df['loan_status'] = df['loan_status'].str.strip().map({'Approved': 1, 'Rejected': 0})


# Fill missing values
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].median(), inplace=True)

# Handle '3+' in Dependents column and convert to numeric
df['no_of_dependents'] = df['no_of_dependents'].replace('3+', '3')
df['no_of_dependents'] = pd.to_numeric(df['no_of_dependents'])

# Encode categorical variables
le = LabelEncoder()
categorical_cols = [col for col in df.columns if df[col].dtype == 'object']
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])


# Features and target
X = df.drop('loan_status', axis=1)
y = df['loan_status']

# Check for NaNs in the target variable after processing
if y.isnull().any():
    print("NaN values still present in target variable after processing.")
    print(y[y.isnull()])

# Scale numerical features (optional but recommended)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# Train Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

print("=== Random Forest Results ===")
print(f"Accuracy: {accuracy_score(y_test, rf_pred):.4f}")
print(classification_report(y_test, rf_pred, zero_division=0))

cm_rf = confusion_matrix(y_test, rf_pred)
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Not Approved', 'Approved'],
            yticklabels=['Not Approved', 'Approved'])
plt.title('Confusion Matrix - Random Forest')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Train XGBoost
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)

print("=== XGBoost Results ===")
print(f"Accuracy: {accuracy_score(y_test, xgb_pred):.4f}")
print(classification_report(y_test, xgb_pred, zero_division=0))

cm_xgb = confusion_matrix(y_test, xgb_pred)
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Greens',
            xticklabels=['Not Approved', 'Approved'],
            yticklabels=['Not Approved', 'Approved'])
plt.title('Confusion Matrix - XGBoost')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Show sample predictions vs actual
results_df = pd.DataFrame({
    'Actual': y_test.reset_index(drop=True),
    'Predicted_RF': rf_pred,
    'Predicted_XGB': xgb_pred
})

print(results_df.head(10))
```
